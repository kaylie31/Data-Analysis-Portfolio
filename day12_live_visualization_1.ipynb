```python
d = pd.read_csv("FoodPrice_in_Turkey.csv")
d.head()
```

```python
# Vẽ bar chart mô tả giá ở từng vùng ('Place') của nhóm sản phẩm 'Rice - Retail' tại thời điểm tháng 12 năm 2019

# Chuẩn bị dữ liệu
# Nhóm sản phẩm 'Rice - Retail' tại tháng 12/2019
# -> Lọc dữ liệu
d1 = d[(d['Month'] == 12) & (d['Year'] == 2019) & (d['ProductName'] == 'Rice - Retail')]

# Bar chart mô tả giá ('Price') ở từng vùng ('Place')
plt.bar(d1['Place'], d1['Price'])
plt.show()
```

```python
# Vẽ line plot (biểu đồ đường) mô tả giá qua từng tháng: x = Month, y = Price
# Lọc điều kiện
  # 1. Place = 'National Average'
  # 2. Year = 2019
  # 3. ProductName = 'Rice - Retail'

d1 = d[(d['Place'] == 'National Average') & (d['Year'] == 2019) & (d['ProductName'] == 'Rice - Retail')]
plt.plot(d1['Month'], d1['Price'])
plt.show()
```

```python
# Vẽ scatter plot để kiểm tra mức độ tương quan về giá ('Price') giữa
# 1. giá sản phẩm 'Fuel (gas) - Retail' và
# 2. giá sản phẩm 'Rice - Retail'
# tại thời điểm năm 2019 và địa điểm là 'National Average'

# chuẩn bị dữ liệu
# lọc năm + lọc địa điểm + lọc sản phẩm
d3 = d[(d['ProductName'] == 'Fuel (gas) - Retail') & (d['Year'] == 2019) & (d['Place'] == 'National Average')]
d4 = d[(d['ProductName'] == 'Rice - Retail') & (d['Year'] == 2019) & (d['Place'] == 'National Average')]

# vẽ dựa trên giá của bảng x, giá của bảng y
plt.scatter(d3['Price'], d4['Price'])
plt.show()
```

```python
# Vẽ biểu đồ mô tả sự thay đổi về Giá theo từng năm của nhóm sản phẩm 'Rice - Retail' tại Istanbul

# Định hình được là muốn vẽ biểu đồ gì -> biểu đồ đường (line plot)
# Chuẩn bị dữ liệu
# Lọc dữ liệu
d5 = d[(d['ProductName'] == 'Rice - Retail') & (d['Place'] == 'Istanbul')]
d6 = d5.groupby('Year')['Price'].mean().reset_index()
d6['Year'] = d6['Year'].astype('str')

# Vẽ
plt.plot(d6['Year'], d6['Price'])
plt.show()
```
